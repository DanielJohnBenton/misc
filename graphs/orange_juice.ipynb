```python
import matplotlib
from matplotlib import pyplot
matplotlib.rcParams["figure.figsize"] = (7, 5)

with pyplot.xkcd(scale=1, length=100, randomness=2):
    pyplot.plot([0, 100], color="xkcd:dark orange")
    
    pyplot.text(0.3, 30, "nice amount of bits")
    pyplot.text(0.6, 60, "bit much bits")
    pyplot.text(0.75, 75, "soup threshold")
    pyplot.text(0.1, 10, "add more bits")
    pyplot.text(1.0, 100, "literally an orange")
    
    pyplot.tick_params(bottom="off", labelbottom="off")
    pyplot.yticks([0, 20, 40, 60, 80, 100], ["0%", "20%", "40%", "60%", "80%", "100%"])
    pyplot.xlabel("Bits in orange juice")
    pyplot.gca().set_ylim([0, 100])
    
    pyplot.show();
```
